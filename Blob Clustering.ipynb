# Example of Clustering algorithm using blobs

This is an example of clustering using K-Means algorithm using blobs. 

### Importing the dataset and requisite libraries 

We use the make blobs dataset available in the "sklearn" library which is a commonly used library in ML and Data Science. We import the KMeans algorithm model from the sklearn library and numpy to use and manipulate arrays.

```python
#import the requisite libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

### Exploratory Data Analysis

```python
#generate the dataset
traindata = make_blobs(n_samples = 400, centers = 5, n_features = 2, cluster_std =1.5, random_state = 40)
```

```python
#Visualize the data
datapoint = traindata[0]
plt.scatter(datapoint[:,0], datapoint[:,1])
```

### K-Means Clustering Model

Implement the model on the data 

```python
kmeans = KMeans(n_clusters = 5)
y_pred = kmeans.fit_predict(datapoint)
```

```python
#Explore the cluster centers
clusters = kmeans.cluster_centers_
print(clusters)
```

### Cluster Visualization

```python
#cluster colors
plt.scatter(datapoint[y_pred==0,0],datapoint[y_pred==0,1],s=80,color='red')
plt.scatter(datapoint[y_pred==1,0],datapoint[y_pred==1,1],s=80,color='blue')
plt.scatter(datapoint[y_pred==2,0],datapoint[y_pred==2,1],s=80,color='yellow')
plt.scatter(datapoint[y_pred==3,0],datapoint[y_pred==3,1],s=80,color='green')
plt.scatter(datapoint[y_pred==4,0],datapoint[y_pred==4,1],s=80,color='pink')

#cluster center pointers
plt.scatter(clusters[0][0],clusters[0][1],marker='+',s=150,color='black')
plt.scatter(clusters[1][0],clusters[1][1],marker='+',s=150,color='black')
plt.scatter(clusters[2][0],clusters[2][1],marker='+',s=150,color='black')
plt.scatter(clusters[3][0],clusters[3][1],marker='+',s=150,color='black')
plt.scatter(clusters[4][0],clusters[4][1],marker='+',s=150,color='black')
```
